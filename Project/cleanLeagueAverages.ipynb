```python
import pandas as pd
import numpy as np
```

Data gathered from https://www.basketball-reference.com/leagues/NBA_stats_per_game.html

```python
raw_df = pd.read_csv('./raw/LeagueAverages.csv', skiprows=1)
raw_df.head()
```

```python
df = raw_df.drop(columns = ['Rk', 'Lg'])
for i in range(df.shape[0]):
    df["Season"].values[i] = df["Season"].iloc[i].split("-")[0][:2] + df["Season"].iloc[i].split("-")[1]
#     df["Ht"].values[i] = int(df["Ht"].iloc[i].split("-")[0]) + df["Season"].iloc[i].split("-")[1]
df["Season"] = pd.to_numeric(df["Season"])
df.loc[df['Season'] == 1900, 'Season'] = 2000
```

```python
df[df['Season']>=1985].to_csv('./modified/cleaned_LeagueAverages.csv', index=False)
```
