### Correct RAPTOR to filter out effects due to small sample size.
Options: weight by minutes played or minute player per game because of RAPTOR with small sample size.

```python
pergame_raw = pd.read_csv('./modified/cleaned_no_tot_PlayerPerGame.csv')
pg_use_cols = [
    'season', 'Player', 'Pos', 'Age', 'team', 'G', 'MP',
    'TRB', 'AST', 'STL', 'BLK', 'PTS', 'player_id'
]
pergame = pergame_raw[pg_use_cols].rename(
    columns = {
        'G'   : '# of games',
        'Pos' : 'Position',
        'MP'  : 'Minutes',
        'TRB' : 'Rebounds',
        'AST' : 'Assists',
        'STL' : 'Steals',
        'BLK' : 'Blocks',
        'PTS' : 'Points'
    })
```

```python
pergame
```

```python
reg_df = rs.merge(pergame, on = ['player_id', 'team', 'season']).drop(columns = ['Player', 'season_type'])
reg_df['2022 salary'] = reg_df['salary'] * (reg_df['2022 sal_cap']/reg_df['salary_cap'])
reg_df.columns
```

```python
max(reg_df['mp'])/(40*58)
```

```python
temp1 = (reg_df['raptor_total']*reg_df['mp']/(40*58))
temp1.hist(bins = 1000)
```

```python
((temp1 - temp1.mean())/temp1.std()).hist(bins =1000)
```

```python
temp2 = (reg_df['raptor_total']*reg_df['Minutes']/48)
temp2.hist(bins=1000)
# plt.xlim(-5,5)
```

```python
((temp2- temp2.mean())/temp2.std()).hist(bins =1000)
```

Filtering by minutes played per game is flawed because there are a series of players that have 42.7 mpg but only played 1 game. As such, we're going to use the total minutes/(40*58). 40 because starters tend to play a maximum of 40 mpg & play at least 58 games to qualify for league awards.

```python
reg_df['fixed_raptor'] = temp1
reg_df['fixed_raptor_zscore'] = ((temp1 - temp1.mean())/temp1.std())
```

### Calculate Z-score of RAPTOR and scaled salary per season:

    * Based on the principle that a top-percentile player should get paid like a top-percentile player.

Scaled salaries form a log-normal distribution.

```python
_ = plt.hist(rs["scaled_salary"], bins = 1000)
for quantile in np.quantile(rs[rs["scaled_salary"]>=0]["scaled_salary"], np.arange(0,1, 0.10)):
    plt.axvline(quantile, c = "Red")
# print(np.mean(rs[rs["normalized salary"]>0]["normalized salary"]))
#plt.xticks(np.arange(0,3500,300))
plt.title('Distribution of Scaled Salaries')
```

```python
rs[rs['scaled_salary'] != 0]['scaled_salary'].quantile(np.arange(0,0.2,0.01))
```

```python
np.log10(rs["scaled_salary"].replace(0,min(rs[rs['scaled_salary'] != 0]['scaled_salary']))).hist(bins=1000)
```

```python
logScaleSal = np.log10(rs["scaled_salary"].replace(0,min(rs[rs['scaled_salary'] != 0]['scaled_salary'])))
# Nomedm
((logScaleSal - np.mean(logScaleSal))/np.std(logScaleSal)).hist(bins=1000)
```

```python
reg_df['scaled_salary'] = reg_df['scaled_salary'].replace(0, min(reg_df[reg_df['scaled_salary'] != 0]['scaled_salary']))
reg_df['log_scaled_salary'] = np.log10(reg_df['scaled_salary'])
reg_df['log_scaled_salary_zscore'] = (reg_df['log_scaled_salary'] - reg_df['log_scaled_salary'].mean())/reg_df['log_scaled_salary'].std()
```

```python
reg_df['log_scaled_salary_zscore'].hist(bins =400)
```

```python
reg_df["Value by Salary"] = reg_df["fixed_raptor_zscore"] - reg_df["log_scaled_salary_zscore"]
reg_df.sort_values(by = "Value by Salary", ascending = False)
```
