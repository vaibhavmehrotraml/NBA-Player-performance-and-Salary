## Import required packages

````python
from nba_api.stats.endpoints import shotchartdetail
from nba_api.stats.static import players as api_players
import json
import requests
import pandas as pd
import time
import traceback
````

## Load players to get ID values

````python
def get_player_id(fullName):
    player_json = api_players.find_players_by_full_name(fullName)
    #player_data = json.loads(player_json.get_json())
    if(len(player_json) == 0):
        return -1
    return player_json[0]['id']
````

## Load teams to get ID values

````python
teams = json.loads(requests.get('https://raw.githubusercontent.com/bttmly/nba/master/data/teams.json').text)
````

````python
def get_team_id(team_name):
    for team in teams:
        if team['abbreviation'] == team_name:
            return team['teamId']
    return -1
````

````python
get_team_id('New York Knicks')
````

````python
get_player_id('LeBron James')
````

````python
player_lookups = pd.read_csv("listOfPlayerLookups.csv")
error_record = {'failed_csv': ["dummy message"], 'printout': ["dummy message"]}
error_df = pd.DataFrame(data=error_record)
error_df.to_csv("errors.csv", index=False)

for i, row in player_lookups.iterrows(): 
    try:
        team_abbrev = row['Tm']
        current_team_id = get_team_id(team_abbrev)
        full_name = row["Player"]
        if full_name[-1] == "*":
            full_name = full_name[0:(len(full_name)-1)]
        current_player_id = get_player_id(full_name)
        current_season = row["Season"][0:5] + row["Season"][7:9]

        #print(full_name)
        #print(current_team_id)
        #print(current_player_id)
        #print(current_season)
        shot_json = shotchartdetail.ShotChartDetail(
                    team_id = current_team_id,
                    player_id = current_player_id,
                    context_measure_simple = 'FGA',
                    season_nullable = current_season,
                    season_type_all_star = 'Regular Season')

        shot_data = json.loads(shot_json.get_json())

        relevant_data = shot_data['resultSets'][0]
        headers = relevant_data['headers']
        rows = relevant_data['rowSet']

        # Create pandas DataFrame
        curry_data = pd.DataFrame(rows)
        curry_data.columns = headers
        curry_data.to_csv("data_files/"+ full_name + "_" + team_abbrev + "_" + current_season + ".csv", index=False)
        print("generated " + full_name + "_" + team_abbrev + "_" + current_season + ".csv")
    except Exception as error:
        failed_csv_name = full_name + "_" + team_abbrev + "_" + current_season + ".csv"
        error_record = {'failed_csv': [failed_csv_name], 'printout': (str(error))}
        error_df = pd.DataFrame(data=error_record)
        error_df.to_csv("errors.csv", mode='a', index=False, header=False)
        print("Failed to generate " + failed_csv_name)
    time.sleep(1)
````
